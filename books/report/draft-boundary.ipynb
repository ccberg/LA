```python
import numpy as np
import pandas as pd
import seaborn as sns

from src.tools.plotter import init_plots, store_sns

init_plots()

points = np.array([(90, 16), (74, 16), (42, 16), (88, 16), (110, 21), (920, 38)])
tvla, dlla = points.transpose()

df = pd.DataFrame({
    "TVLA $t$-test, $\mu_1$, min-$p$": tvla,
    "DLLA (Wegener)": dlla
}).set_index(np.arange(0, 3, .5))

G = sns.lineplot(data=df)
G.set(title="Leakage boundaries for unprotected AES,\nunder the jitter measurement error",
                          xlabel="Jitter parameter", ylabel="Leakage boundary")

store_sns(G, "aisy-leakage-bound")
```
