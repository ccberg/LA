```python
import numpy as np
import statsmodels.api as sm
import pandas as pd
import seaborn as sns

np.random.seed(42)

ar = np.r_[1, -np.array([3/4, -1/4])]
ma = np.r_[1, np.array([2/3, 1/3])]
y = sm.tsa.arma_generate_sample(ar, ma, 1400)

y_prime = y * 16

sns.lineplot(data=pd.DataFrame(zip(y, y_prime), columns=["ARMA", "Trace"]))
```
