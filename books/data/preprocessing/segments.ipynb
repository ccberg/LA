# Segment extractor

Chops the ASCAD raw database up with a window of 1400, step 700

```python
from more_itertools import windowed

import h5py
from src.aes.main import AES
from src.data.loaders.ascad import ASCADData, ASCAD_DATA, ASCAD_DATA_VAR, ASCADDataType, ASCAD_DATA_FIX
from tqdm import tqdm
import numpy as np

STEP = 700
TRACE_LEN  = 1400
IX_CLIP = 80945
```

```python
TRACES = h5py.File(f"/home/berg/ASCAD_raw.h5", 'r')['traces']
WINDOWS = np.array(list(windowed(range(len(TRACES)), n=TRACE_LEN + 1, step=STEP)))[:-1]
```

```python
np.where(WINDOWS[:, 0] > IX_CLIP)[0][0]
```

```python
# wrt = h5py.File(f"{ASCAD_DATA}/ASCAD_data/ASCAD_databases/raw-segments.h5", 'w-')
# wrt.close()
```

```python
num_traces = TRACES.shape[0]
for win_ix, window in enumerate(WINDOWS):
    wrt = h5py.File(f"{ASCAD_DATA}/ASCAD_data/ASCAD_databases/raw-segments.h5", 'r+')

    if str(win_ix) in wrt:
        continue

#     if not 115 <= win_ix <= 130:
#         continue

    a, b = window[0], window[-1]

    acc = np.zeros((num_traces, TRACE_LEN), dtype=np.uint8)
    for ix in tqdm(range(num_traces), desc=f"Window ix: {win_ix}"):
        acc[ix] = TRACES[ix][a:b]

    wrt[str(win_ix)] = acc

    wrt.close()
```
